```python
#数据对象相关性
attribute=["price"]
t4[attribute] = t4[attribute].fillna(0)

# 对数据进行正则化
t4[attribute] = t4[attribute].apply(lambda x : (x - np.mean(x)) / (np.max(x) - np.min(x)))

# 构造分数表
score = {}
range_length = len(t4)
for i in range(0, range_length):
    score[i] = {}
    for j in range(0, range_length):
        score[i][j] = 0    


for i in range(0, range_length):
    for j in range(i, range_length):
        for item in attribute:
            temp = abs(t4.iloc[i][item] - t4.iloc[j][item])
            score[i][j] += temp
        score[j][i] = score[i][j]
```

```python
# 对有缺失值的条目，用和它相似度最高（得分最低）的数据条目中对应属性的值替换
nan_list = pd.isnull(t4).any(1).nonzero()[0]
for index in nan_list:
    best_friend = sorted(score[index].items(), key=operator.itemgetter(1), reverse = False)[1][0]
    for item in attribute:
        if pd.isnull(t4.iloc[index][item]):
            if pd.isnull(train.iloc[best_friend][item]):
                t4.ix[index, item] = train[item].value_counts().idxmax()
            else:
                t4.ix[index, item] = train.iloc[best_friend][item]
```
